```python
from urllib.request import urlopen
url = 'https://en.wikipedia.org/wiki/Blog'
html_content = urlopen(url).read()
print(html_content)
```

# data cleaning

```python
'''
Beautiful Soup is a Python library for pulling data out of HTML and XML files. 
It works with your favorite parser to provide idiomatic ways of navigating, searching, and modifying the parse tree. 
It commonly saves programmers hours or days of work.
'''

from bs4 import BeautifulSoup
html_content = BeautifulSoup(html_content)
print(html_content)
```

# only text, no style and script

```python
text = html_content.get_text()
print(text)
```

# now will remove spaces before and after of the lines

```python
'''
The Strip() method in Python removes or truncates the given characters from the 
beginning and the end of the original string. The default behavior of the strip() 
method is to remove the whitespace from the beginning and at the end of the string
'''

lines = []
for line in text.splitlines():
  lines.append(line.strip())


lines = tuple(lines)
print(lines)
```

# now let's remove extra spaces in between the words


```python
words = []
for line in lines:
    for word in line.split(" "):
        words.append(word.strip())
```

# now its time to make content


```python
word = []
for i in words:
    if i.isalpha():
        word.append(i)

content = " ".join(word)
print("This is my content :-\n")
content
```

```python
import nltk
#nltk.download('all')
#nltk.download('stopwords')
#nltk.download('punkt')
#nltk.download('wordnet')

''' 
punkt;-
This tokenizer divides a text into a list of sentences by using an unsupervised 
algorithm to build a model for abbreviation words, collocations, and words that start sentences.

wordnet:
WordNet alongside the NLTK module to find the meanings of words, synonyms, antonyms,
'''
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import SnowballStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
content = content.lower()
content = word_tokenize(content)
print("Content after word tokenization \n",content)
```

# Stemming/Lemmatization

```python
lemmatizer = WordNetLemmatizer()
word = []
for i in content:
    words = lemmatizer.lemmatize(i)
    word.append(words)
content = ' '.join(word)
print("this the content after tokenization and lemmatization \n",content)
```

# stop words

```python
'''
in case if you want to add/remove any stop words:-
stopwords = nltk.corpus.stopwords.words('english')
stopwords.append('newWord')
stopwords.remove('newWord')
'''

stpwords = set(stopwords.words('english'))
print("type is :-\n",type(stopwords.words('english')))
print("stop words in english:-\n",stpwords)
print("stop words length in english:-\n",len(stpwords))
word = []
for i in content.split():
    if i not in stpwords:
        word.append(i)

content = " ".join(word)
print('Content after tokenization, Lemmatization and Stop words:- \n', content)
```

# part of speech:- verb, adjective, noun, etc


```python
from nltk import pos_tag
content = list(content.split())
content = pos_tag(content)
print("This is how content will look like after POS:-\n",content)
```

# Name entity Recognition

```python
word = []
for key, value in content:
    word.append(key)
content = ' '.join(word)

import spacy
nlp = spacy.load("en_core_web_sm")
content = nlp(content)
for ent in content.ents:
    print(ent.text, "|", ent.label_, "|", spacy.explain(ent.label_))

from spacy import displacy
# content = nlp("Tesla Inc is going to acquire twitter for $45 billion")
display = displacy.render(content, style="ent")
print(display)
```
